```python
import trimesh
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# OFF 파일 불러오기
mesh = trimesh.load_mesh('/content/ModelNet40/airplane/train/airplane_0005.off')

# 3D 시각화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 삼각형 면을 추출하여 그리기
triangles = [mesh.vertices[face] for face in mesh.faces]
ax.add_collection3d(Poly3DCollection(triangles, facecolors='lightblue', linewidths=0.5, edgecolors='gray', alpha=0.5))

# 축설정 및 시각화
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.auto_scale_xyz(mesh.bounds[:, 0], mesh.bounds[:, 1], mesh.bounds[:, 2])
plt.show()
```
